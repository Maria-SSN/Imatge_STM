**Import libraries**

```python
from sisl import *
import numpy as np
```

**Import molecule in xyz format and center using sisl**

We can use the center() to find the where is the center now center and simply add a vector to translate the system to the desired place
Documentation on geometry manipulation here:

https://sisl.readthedocs.io/en/latest/api/generated/sisl.Geometry.html

```python
system=Geometry.read("molecule.xyz")
system.plot()
```

```python
#where is the center now
system_center=system.center()
print(system_center)
```

*vector with the new center coordinates*

```python
new_center=np.array([7,7,5])
```

*move the coordinates to the new place*

```python
system.xyz=system.xyz-system_center+new_center
```

```python
system.plot()
```

**Convert to fdf format**

```python
system.write("structure.fdf")
```
